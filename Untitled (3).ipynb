```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# 3Blue1Brown-ish color theme
BG_COLOR      = "#0b1020"  # dark navy
ACCENT_BLUE   = "#1f77b4"
ACCENT_ORANGE = "#ff9933"
ACCENT_GREEN  = "#00cc88"
ACCENT_PURPLE = "#aa66ff"
TEXT_COLOR    = "#f5f5f5"
GRID_COLOR    = "#444444"

plt.style.use("default")
plt.rcParams.update({
    "figure.facecolor": BG_COLOR,
    "axes.facecolor": BG_COLOR,
    "axes.edgecolor": TEXT_COLOR,
    "axes.labelcolor": TEXT_COLOR,
    "xtick.color": TEXT_COLOR,
    "ytick.color": TEXT_COLOR,
    "text.color": TEXT_COLOR,
    "grid.color": GRID_COLOR,
    "axes.grid": True,
    "grid.alpha": 0.3,
    "font.size": 12,
})
```

```python
# Make sure iris_measurements.csv is in the same folder as the notebook
df = pd.read_csv("iris_measurements.csv")
df.head()
```

```python
from itertools import combinations

X = df.values
feature_names = df.columns.tolist()

pairs = list(combinations(range(4), 2))

fig, axes = plt.subplots(2, 3, figsize=(12, 8), facecolor=BG_COLOR)
axes = axes.ravel()

for ax, (i, j) in zip(axes, pairs):
    ax.scatter(X[:, i], X[:, j], s=30, color=ACCENT_BLUE, alpha=0.7)
    ax.set_xlabel(feature_names[i])
    ax.set_ylabel(feature_names[j])

for k in range(len(pairs), len(axes)):
    axes[k].axis("off")

fig.suptitle("Iris measurements (no answer key)", color=TEXT_COLOR, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

fig, ax = plt.subplots(figsize=(7, 6))
ax.scatter(X_pca[:, 0], X_pca[:, 1],
           s=35, color=ACCENT_BLUE, alpha=0.85)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title("Unlabeled structure in 2D")
plt.tight_layout()
plt.show()
```

```python
cluster_colors = [ACCENT_ORANGE, ACCENT_GREEN, ACCENT_PURPLE, "#ff4444"]

fig, axes = plt.subplots(1, 3, figsize=(15, 4), facecolor=BG_COLOR)
ks = [2, 3, 4]

for ax, k in zip(axes, ks):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = kmeans.fit_predict(X_scaled)

    for cid in range(k):
        mask = labels == cid
        ax.scatter(
            X_pca[mask, 0],
            X_pca[mask, 1],
            s=40,
            color=cluster_colors[cid],
            alpha=0.9
        )

    centers_2d = pca.transform(kmeans.cluster_centers_)
    ax.scatter(
        centers_2d[:, 0],
        centers_2d[:, 1],
        s=160,
        color=BG_COLOR,
        edgecolor=TEXT_COLOR,
        linewidth=2,
        marker="o"
    )

    ax.set_title(f"k = {k}")
    ax.set_xticks([])
    ax.set_yticks([])

fig.suptitle("Unsupervised groupings (no species column)", color=TEXT_COLOR, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
labels = kmeans.fit_predict(X_scaled)

fig, ax = plt.subplots(figsize=(7, 6))

for cid, color in zip(range(3), [ACCENT_ORANGE, ACCENT_GREEN, ACCENT_PURPLE]):
    mask = labels == cid
    ax.scatter(
        X_pca[mask, 0],
        X_pca[mask, 1],
        s=45,
        color=color,
        alpha=0.9
    )

ax.set_xticks([])
ax.set_yticks([])
ax.set_title("Many measurements, no given species\n unsupervised clusters", fontsize=14)
plt.tight_layout()
plt.show()
```
